```python
from bokeh.charts import Histogram
# import charts as you use them
from bokeh.sampledata.iris import flowers as data

hist = Histogram(data, values = 'petal_length', legend = 'top_right', bins = 12, color = 'species')

# bins defines how many groups the data will be divded into.
# color is like a pivot and can be grouped

show(hist)
```
